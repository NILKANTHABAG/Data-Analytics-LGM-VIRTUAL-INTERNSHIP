# Task 4 - Image to Pencil Sketch

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from PIL import Image
from IPython.display import display
```

```python
img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Untitled.png", 1) 
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, (640, 400))
display(Image.fromarray(img))
```

# **Converting** image to grey scale

```python
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_grey = cv2.resize(img_grey, (640, 400))
display(Image.fromarray(img_grey))
```

## **Inverting the image**

```python
img_inv = cv2.bitwise_not(img_grey)
img_inv = cv2.resize(img_inv, (640, 400))
display(Image.fromarray(img_inv))
```

```python
img_smooth = cv2.GaussianBlur(img_inv, (5, 5), 0, 0)
img_smooth = cv2.resize(img_smooth, (640, 400))
display(Image.fromarray(img_smooth))
```

## **Inverting the image to get final sketch**

```python
def final(img_grey, img_smooth):
  img_smooth_inv = 255 - img_smooth
  return cv2.divide(img_grey, img_smooth_inv, scale=256.0)
```

```python
img_final = final(img_grey, img_smooth)
img_final = cv2.resize(img_final, (640, 400))
display(Image.fromarray(img_final))
```

# **Comparing the Original Image with Pencil Sketch**

```python
display(Image.fromarray(img))
display(Image.fromarray(img_final))
```
